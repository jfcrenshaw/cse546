```python
x1 = np.linspace(-10,10)
x2 = 1/2 * x1 - 1

fig,ax = plt.subplots(figsize=(3.1,3),constrained_layout=True)

ax.plot(x1,x2)

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_yticks([-10,-5,0,5,10])

fig.savefig('A9a.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

x1 = np.linspace(-10,10)
x2 = np.linspace(-10,10)
x1grid, x2grid = np.meshgrid(x1,x2)
x3grid = -x1grid - x2grid

fig = plt.figure(constrained_layout=True)
ax = plt.axes(projection='3d')

ax.contour3D(x1grid, x2grid, x3grid, 300)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.set_zlabel("$x_3$")
ax.dist=11

fig.savefig('A9b.pdf')
```

```python
import numpy as np

A = np.array([[0, 2, 4], [2, 4, 2], [3, 3, 1]])
invA = np.linalg.inv(A)
b = np.array([-2, -2, -4])
c = np.array([1, 1, 1])

print("A^{-1}b")
print(invA @ b)

print("\nAc")
print(A @ c)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(11)

n = int(1.6e5)

fig,ax = plt.subplots(figsize=(4,3.4))

colors = ['C0','C2','C3','C4']
for i,k in enumerate([1,8,64,512]):
    Y = np.sum(np.sign(np.random.randn(n, k))*np.sqrt(1/k), axis=1)
    plt.plot(sorted(Y), np.arange(1,n+1)/float(n), label=k, c=colors[i])
    
Z = np.random.randn(n)
plt.step(sorted(Z), np.arange(1,n+1)/float(n), label='Gaussian', c='C1')

plt.legend()
plt.xlim(-3,3)
plt.ylim(0,1)
plt.xlabel('Observations')
plt.ylabel('Probability')

fig.savefig('A12.pdf')
```
